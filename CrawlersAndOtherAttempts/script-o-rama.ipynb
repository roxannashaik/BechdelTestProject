```python
import requests
from bs4 import BeautifulSoup
import lxml.html as lh
import PyPDF2
import urllib.request  as urllib2
import webbrowser
```

```python
import pandas as pd
```

```python
def get_urls():
    urls = ['http://www.script-o-rama.com/snazzy/table.html','http://www.script-o-rama.com/snazzy/table2.html',
           'http://www.script-o-rama.com/snazzy/table3.html','http://www.script-o-rama.com/snazzy/table4.html']
    res = []
    for url in urls:
        req = requests.get(url)
        soup = BeautifulSoup(req.content, 'lxml')
        a = soup.find_all('a', href=True)
       # b = soup.find_all('a').contents[0]
        for item in a:
            if 'http' in item['href']:
                res.append((item.contents[0], item['href']))
    return res
urls = get_urls()
```

```python
def parse(urls):
    c1 = 0
    c2 = 0
    c3 = 0
    print(len(urls))
    result = []
    for url1 in urls:
        url = url1[1]
        end = url.split(".")[-1]
        try:
            if(end == 'html'):
                c1 +=1
                #print('html')
                req = requests.get(url, timeout=5)
                soup = BeautifulSoup(req.content,'lxml')
                result.append((url1[0],soup.text))
            if(end == "pdf"):
                print(url1[0])
                c2 +=1
                #print('pdf')
                #d = download_file(url)
                #data = []
                #with open(d, 'rb') as G:
                    #print(G.read())
                   # data = G.read()
                   # print(data)
               # print(i.read())
                #pdfReader = PyPDF2.PdfFileReader(i) 
                #result.append(data)
            if(end == "txt"):
                c3 +=1
                #print('txt')
                data = urllib2.urlopen(url).read()
                result.append((url1[0],data))
        except:
            print(url)
    print(c1)
    print(c2)
    print(c3)
    return result
res = parse(urls)
```
